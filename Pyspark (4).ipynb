```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://www.niit.com/india/placement-linked-programs'

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Initialize lists to store the extracted data
program_names = []
program_details = []
program_modes = []

# Locate all program blocks (You may need to adjust the class based on actual page structure)
program_blocks = soup.find_all('div', class_='prog-text')

for program in program_blocks:
    # Extract program name
    name = program.find('h4', class_='courname').get_text(strip=True)
    program_names.append(name)
    # Extract program details like duration, mode, etc.
    details = program.find_all('span', class_='text-prog')
    detail_texts = [detail.get_text(strip=True) for detail in details]
    program_details.append(', '.join(detail_texts))
    # Extract mode of delivery
    mode = program.find('span', class_='mode-delivery').get_text(strip=True) if program.find('span', class_='mode-delivery') else "N/A"
    program_modes.append(mode)

# Create a DataFrame to organize the data into a table
df = pd.DataFrame({
    'Program Name': program_names,
    'Details': program_details,
    'Mode of Delivery': program_modes
})

# Display the DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the webpage to scrape
url = 'https://www.niit.com/india/placement-linked-programs'

# Send a GET request to the website
response = requests.get(url)

# Parse the HTML content of the page with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Initialize lists to store the extracted data
program_names = []
durations = []
modes = []
special_notes = []

# Locate all program blocks (You may need to adjust the class based on actual page structure)
program_blocks = soup.find_all('div', class_='prog-text')

for program in program_blocks:
    # Extract program name
    name = program.find('h4', class_='courname').get_text(strip=True)
    program_names.append(name)
    # Extract details
    details = program.find_all('span', class_='text-prog')
    duration = "N/A"
    mode = "N/A"
    special_note = "N/A"
    for detail in details:
        text = detail.get_text(strip=True)
        if "Duration" in text:
            duration = text
        elif "Full-Time" in text or "Part-Time" in text:
            mode = text
        else:
            special_note = text
    durations.append(duration)
    modes.append(mode)
    special_notes.append(special_note)

# Create a DataFrame to organize the data into a table
df = pd.DataFrame({
    'Program Name': program_names,
    'Duration': durations,
    'Mode': modes,
    'Special Notes': special_notes
})

# Display the DataFrame
print(df)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the NIIT Placement Linked Programs page
url = "https://www.niit.com/india/placement-linked-programs"

# Send a GET request to fetch the content of the page
response = requests.get(url)
response.raise_for_status()  # Ensure we got a valid response

# Parse the content using BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Initialize an empty list to store program details
programs = []

# Locate the sections containing the program details
sections = soup.find_all('div', class_='program-section')  # Update the class based on the actual HTML structure

# Loop through each section and extract the required details
for section in sections:
    program_name = section.find('h3').get_text(strip=True) if section.find('h3') else "-"
    duration = section.find('span', class_='duration').get_text(strip=True) if section.find('span', class_='duration') else "-"
    mode_of_learning = section.find('span', class_='mode').get_text(strip=True) if section.find('span', class_='mode') else "-"
    placement_assistance = section.find('span', class_='placement').get_text(strip=True) if section.find('span', class_='placement') else "-"
    additional_features = section.find('span', class_='features').get_text(strip=True) if section.find('span', class_='features') else "-"

    # Append the extracted details to the programs list
    programs.append({
        "Program Name": program_name,
        "Duration": duration,
        "Mode of Learning": mode_of_learning,
        "Placement Assistance": placement_assistance,
        "Additional Features": additional_features
    })

# Convert the list of dictionaries to a DataFrame
df = pd.DataFrame(programs)

# Display the DataFrame
df
```

```python
# Extract duration in weeks from the "Details" column
#spark2_df = spark_df.withColumn("Duration_Weeks", F.regexp_extract("Details", r'(\d+)\s+Weeks', 1).cast("int"))

# Extract part-time/full-time information from the "Details" column
spark2_df = spark_df.withColumn("Time_Type",
                   F.when(spark_df.Details.like('%Part-Time%'), 'Part-Time')
                   .when(spark_df.Details.like('%Full-Time%'), 'Full-Time')
                   .otherwise(None))

# Extract online/offline information from the "Details" column
spark2_df = spark2_df.withColumn("Mode",
                   F.when(spark2_df.Details.like('%Online%'), 'Online')
                   .when(spark2_df.Details.like('%Offline%'), 'Offline')
                   .when(spark2_df.Details.like('%Virtual Learning%'), 'Virtual Learning')
                   .when(spark2_df.Details.like('%Blended Learning%'), 'Blended Learning')
                   .when(spark2_df.Details.like('%Instructor Led Face to Face%'), 'Instructor Led Face to Face')
                   .otherwise(None))

# Show the modified DataFrame
spark2_df.show()
```

```python
from pyspark.sql.functions import when  # Import 'when' function from pyspark.sql.functions
spark2_df = spark2_df.withColumn("Mode", when(spark2_df.Mode == "Instructor Led Face to Face", "Offline").otherwise(spark2_df.Mode))

# Replace 'Virtual Learning' with 'Offline'
spark2_df = spark2_df.withColumn("Mode", when(spark2_df.Mode == "Virtual Learning", "Online").otherwise(spark2_df.Mode))
```
